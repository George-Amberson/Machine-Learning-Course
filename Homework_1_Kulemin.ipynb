По данным о среднем, разбросе а также минимуме и максимуме можно предположить наличие выбросов, используя метод межквартильного диапазона, позже удалим их:

### 3. Обработка категориальных признаков

Закодируем категориальный признак используя встроенный в scikit-learn модуль

```python
ord_enc = OrdinalEncoder()
data["Sector"] = ord_enc.fit_transform(data[["Sector"]])
```

### 4. Обработка выбросов и пропущенных значений

Удалим выбросы из колонок используя метод межквартильного интервала

```python
Q1 = data.quantile(q=.25)
Q3 = data.quantile(q=.75)
IQR = data.apply(stats.iqr)
num_data_clean = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
num_data_clean.shape
```

Определим наличие незаполненных значений

```python
columns_with_more_then_85_percents_missings =[]
for col in num_data_clean:
    pct_missing = np.mean(num_data_clean[col].isnull())
    if(pct_missing*100>0): print('{} - {}%'.format(col, (pct_missing*100)))
```

Заполним пропущенные значения используя среднее по каждому столбцу

```python
num_data_clean = num_data_clean.fillna(num_data_clean.mean())
columns_with_more_then_85_percents_missings =[]
for col in num_data_clean:
    pct_missing = np.mean(num_data_clean[col].isnull())
    if(pct_missing*100>0): print('{} - {}%'.format(col, (pct_missing*100)))
```

### 5. Нормализация данных

Применим к значениям нормализацию минимумом-максимумом



```python
def normalize(df):
    result = df.copy()
    max_value = df.max()
    min_value = df.min()
    result = (df - min_value) / (max_value - min_value)
    return result
num_data_clean.apply(lambda x: x if np.std(x) == 0 else normalize(x))
```

### 6. Разбиение данных на обучающие и тестовые

Разделим данные на тренировочные и тестовые

```python
X, y = num_data_clean.drop(['Class'], axis=1), num_data_clean['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    train_size=0.67, 
                                                    random_state=42)
```

### 6. Запустим классификатор ближайших соседей с разным колличеством классов

Построим классификатор "К ближайших соседей" при этом проэксперементируем с оптимальным числом классов

```python
from sklearn.neighbors import KNeighborsClassifier
accuracy_list_train = []
accuracy_list_test = []
for i in range(1, 45):
    KNN = KNeighborsClassifier(n_neighbors=i)
    KNN.fit(X_train, y_train)
    
    accuracy_train = round(KNN.score(X_train, y_train)*100, 4)
    accuracy_list_train.append(accuracy_train)
    
    accuracy_test = round(KNN.score(X_test, y_test)*100, 4)
    accuracy_list_test.append(accuracy_test)
```

### 7. Вычислим ошибки на обучающей и тестовой выборке

Построим график средней точности на обучающей выборке

```python
k = np.arange(1, 45)
plt.figure(figsize=(12, 6))
plt.plot(k, accuracy_list_train, 'red')
plt.xlabel('k')
plt.xticks(np.arange(min(k), max(k)+1, 1.0))
plt.grid(which='major')
plt.ylabel('Accuracy on train, %')
```

Построим график средней точности на тестовой выборке

```python
k = np.arange(1, 45)
plt.figure(figsize=(12, 6))
plt.plot(k, accuracy_list_test, 'blue')
plt.xlabel('k')
plt.xticks(np.arange(min(k), max(k)+1, 1.0))
plt.grid(which='major')
plt.ylabel('Accuracy on test, %')
```

Нетрудно определить что оптимальным числом классов будет 31. Построим классификатор на этом числе классов и оценим его матрицу ошибок а также взвешенную f1 оценку

```python
from sklearn.metrics import confusion_matrix
knn = KNeighborsClassifier(n_neighbors = 31)
knn.fit(X_train, y_train)
y_test_predict = knn.predict(X_test)
print(confusion_matrix(y_test, y_test_predict))
from sklearn.metrics import f1_score
f1_score(y_test, y_test_predict, average='weighted')
```

### 8. Сравним классификатор ближайших соседей с RandomForestClassifier

Результат можно интепретировать как 70% точность классификатора, что является неплохим но вполне ожидаемым результатом для такого достаточно простого классификатора. Для сравнения применим классификатор RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=8, random_state=0)
clf.fit(X_train,  y_train)
y_test_predict = clf.predict(X_test)
from sklearn.metrics import confusion_matrix
error_train = np.mean(y_train != clf.predict(X_train))
error_test  = np.mean(y_test  != clf.predict(X_test))
print('Error on train = ', error_train)
print('Error on test = ', error_test)
print(confusion_matrix(y_test, y_test_predict) )
from sklearn.metrics import f1_score
f1_score(y_test, y_test_predict, average='weighted')
```

Разница видна невооруженным взглядом, RandomFoest дает почти 95% точность
